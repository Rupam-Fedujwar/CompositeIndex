```python
# Expert aggregation based on role
expert = df.copy()
expert['count'] = 1
expert_role =  expert.groupby(['Your role'])['count'].sum()

# Data to plot
labels = expert_role.index
sizes = expert_role.values  # These values represent the percentages for each category
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'pink']
explode = (0, 0.2, 0, 0, 0)  # Explode a slice (i.e., 'B')

# Create the pie chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140, textprops={'fontsize': 18})
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Display the pie chart
plt.show()
```

```python
# Plot a bar chart for gender distribution with rotated x-axis labels and percentages
plt.figure(figsize=(8, 6))
gender_counts = expert['Gender'].value_counts()
gender_percentage = gender_counts / gender_counts.sum() * 100
gender_counts.plot(kind='bar', color=['yellowgreen', 'lightcoral', 'green', 'purple'])
# plt.xlabel('Gender')
#plt.ylabel('Number of experts', fontsize=14)

# Remove the grid
plt.grid(False)

# Rotate x-axis labels by 45 degrees
plt.xticks(rotation=0, fontsize=16)
#plt.yticks(rotation=0, fontsize=16)

# Remove y-ticks
plt.yticks([])

# Add percentages above the bars with increased font size
for i, count in enumerate(gender_counts):
    plt.text(i, count + 0.5, f'{gender_percentage[i]:.1f}%', ha='center', fontsize=16)

plt.show()
```

```python
# Calculate experience category percentages
experience_percentages = expert['Experience in above stated Role'].value_counts()

# Sort the index in the desired order
experience_percentages = experience_percentages.loc[['5 years or Less', 'Between 5-10 years', 'Between 10-15 years', 'Between 15-20 years', '20 Years or More']]

# Plot line plot for experience category percentages
plt.figure(figsize=(8, 6))
experience_percentages.plot(marker='o',  linestyle='-', color='b', markersize=8)

# Remove the grid
plt.grid(False)

# Set y-axis limit starting from 0
plt.ylim(0, experience_percentages.max() + 5)

# plt.title('Experience Category Distribution')
# plt.xlabel('Experience Category')
plt.ylabel('Number of experts', fontsize=14)
plt.xticks(rotation=45,fontsize=16)
# plt.grid(True)
plt.show()
```
